```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
games = pd.read_csv('/content/drive/MyDrive/kaggle/archive/games.csv')

games.head()
```

```python
games.info()
```

```python
games=games[['GAME_DATE_EST', 'GAME_ID', 'GAME_STATUS_TEXT','TEAM_ID_home','TEAM_ID_away','PTS_home', 'PTS_away', 'HOME_TEAM_WINS']]
games.head()
```

```python
# Cambio de tipo de dato en la columna GAME_DATE_EST

games['GAME_DATE_EST']= pd.to_datetime(games['GAME_DATE_EST'])
```

```python
games.info()
```

```python
games.head()
```

```python
# Nueva columna con el año de los partidos jugados
games['year']=games['GAME_DATE_EST'].dt.year
games.head()
```

```python
# Cargar el conjunto de teams
teams = pd.read_csv('/content/drive/MyDrive/kaggle/archive/teams.csv')
teams.head()
```

```python
teams=teams[['TEAM_ID', 'CITY', 'NICKNAME']]
teams.head()
```

```python
home_games = pd.merge(games,teams, left_on='TEAM_ID_home', right_on='TEAM_ID', how='inner')
home_games.head()
```

```python
home_games=home_games.rename(columns={
    'CITY' : 'city_home',
    'NICKNAME' : 'nickname_home'
})
home_games.head()
```

```python
both_games = pd.merge(home_games,teams, left_on='TEAM_ID_away', right_on='TEAM_ID', how='inner')
both_games.head()
```

```python
both_games=both_games.rename(columns={
    'CITY' : 'city_away',
    'NICKNAME' : 'nickname_away'
})
home_games.head()
```

```python
both_games=both_games.drop(columns = ['TEAM_ID_home', 'TEAM_ID_away'])
both_games.head()
```

```python
both_games=both_games.drop(columns = ['TEAM_ID_x','TEAM_ID_y'])
both_games.head()
```

```python
games=both_games.copy()
games.columns=games.columns.str.lower()
games.head()
```

## ¿Se han ganado más partidos estando en casa o de visitante?

```python
games['total_points']=games['pts_home']+games['pts_away']
games.head()
```

```python
games.nlargest(n=5, columns='total_points')
```

```python
games.nsmallest(n=5, columns='total_points')
```

```python

total = games['home_team_wins'].value_counts()
total
```

```python
plt.pie(total.values, labels=['Casa', 'Visitante'], autopct='%.0f%%')
plt.show()
```

##¿Qué equipos han hecho la mayor cantidad de puntos (tanto en casa como visitantes)?

###En casa

```python
plt.figure(figsize=(20,6))
plt.xticks(rotation=75)
home_wins=games[games['home_team_wins']== 1]
result=home_wins.groupby('nickname_home')['pts_home'].aggregate(np.mean).reset_index().sort_values('pts_home')

ax=sns.barplot(data=home_wins, x='nickname_home', y='pts_home', order=result['nickname_home'], errorbar=None)
ax.bar_label(ax.containers[-1], fmt='%.0f', label_type='center', rotation=75)
plt.show()
```

###Visitantes


```python
plt.figure(figsize=(20,6))
plt.xticks(rotation=75)
away_wins=games[games['home_team_wins']== 0]
result=home_wins.groupby('nickname_away')['pts_away'].aggregate(np.mean).reset_index().sort_values('pts_away')

ax=sns.barplot(data=away_wins, x='nickname_away', y='pts_away', order=result['nickname_away'], errorbar=None)
ax.bar_label(ax.containers[-1], fmt='%.0f', label_type='center', rotation=75)
plt.show()
```

##¿Contra qué equipos se ha ganado más estando en casa y de visitante?

```python
x = home_wins[['nickname_home','nickname_away']].value_counts()

ax=x[:10].plot(kind='bar', figsize=[10,5])
ax.set_title('Casa vs Visita')
ax.set_xlabel('Partidos')
ax.set_ylabel('Cantidad')
ax.set_xticklabels(x.index[:10], rotation=90)

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom')

plt.show()
```

```python
x = away_wins[['nickname_home','nickname_away']].value_counts()

ax=x[:10].plot(kind='bar', figsize=[10,5])
ax.set_title('Casa vs Visita')
ax.set_xlabel('Partidos')
ax.set_ylabel('Cantidad')
ax.set_xticklabels(x.index[:10], rotation=90)

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom')

plt.show()
```

## 10 jugadores que hiceron más puntos en el año 2022

```python
game_details=pd.read_csv('/content/drive/MyDrive/kaggle/archive/games_details.csv')
game_details.head()
```

```python
game_details.info()
```

```python
game_details=game_details[['GAME_ID','PLAYER_NAME','PTS']]
game_details.head()
```

```python
games_complete=pd.merge(game_details, games, left_on='GAME_ID', right_on='game_id',how='inner')
games_complete.head()
```

```python
player_pts = games_complete.groupby(['year', 'PLAYER_NAME'])['PTS'].mean().reset_index().sort_values('year')
player_pts.head()
```

```python
player_pts[player_pts['year'] == 2022].sort_values(by='PTS', ascending=False).head(10)
```
